Given a number n, find the nth number in the Padovan Sequence.

A Padovan Sequence is a sequence which is represented by the following recurrence relation

P(n) = P(n-2) + P(n-3)

P(0) = P(1) = P(2) = 1

Note: Since the output may be too large, compute the answer modulo 10^9+7.

Examples :

Input: n = 3

Output: 2

Explanation: We already know, P1 + P0 = P3 and P1 = 1 and P0 = 1

Input: n = 4

Output: 2

Explanation: We already know, P4  = P2 + P1 and P2 = 1 and P1 = 1

Expected Time Complexity: O(n)

Expected Auxiliary Space: O(1)

Constraints:

1 <= n <= 106

```python
#User function Template for python3

class Solution:
    def padovanSequence(self,n):
        # code here 
        if n<=2:
            return 1 
        elif n == 3 or n==4 :
            return 2
        mod = 1000000007
        a, b, c = 1,2,2
        result = 0
        for i in range(5,n+1):
            result  = (a + b)%mod
            a , b, c = b, c, result
        
        return result
            


#{ 
 # Driver Code Starts
#Initial Template for Python 3

if __name__ == '__main__':
    t = int(input())
    for _ in range(t):
        n = int(input())
        ob = Solution()
        print(ob.padovanSequence(n))

# } Driver Code Ends    
```
